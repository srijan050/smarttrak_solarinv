```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('dataSample1.csv')
df.head()
```

```python
eff_act = df['Efficiency (%)'].values
amb_temp = df['Ambient Temperature(° C)'].values

for i in range(len(eff_act)):
    if eff_act[i] >= 300:
        print(i)
```

```python
# Plot temperature vs efficiency
plt.figure(figsize=(8,6))
plt.plot(amb_temp[:1000], eff_act[:1000], 'o', label='Efficiency vs Temperature')

# Graph labels
plt.title('Efficiency vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Efficiency (%)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
igbt_temp = df['IGBT Temperature(° C)'].values

# Plot temperature vs efficiency
plt.figure(figsize=(8,6))
plt.plot(igbt_temp[:1000], eff_act[:1000])

# Graph labels
plt.title('Efficiency vs IGBT Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Efficiency (%)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
heatsink_r_temp = df['Heatsink R Temperature(° C)'].values
heatsink_s_temp = df['Heatsink S Temperature(° C)'].values
heatsink_t_temp = df['Heatsink T Temperature(° C)'].values

# Plot temperature vs efficiency
plt.figure(figsize=(8,6))
plt.plot(heatsink_r_temp[:1000], eff_act[:1000])
plt.plot(heatsink_s_temp[:1000], eff_act[:1000])
plt.plot(heatsink_t_temp[:1000], eff_act[:1000])

# Graph labels
plt.title('Efficiency vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Efficiency (%)')
plt.grid(True)

# Show the plot
plt.show()
```
